### 1. What is name of car that has the highest horsepower?

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
#your code here
# Filter cars matching the conditions
filtered_cars = df[(df['horsepower'] > 100) & (df['weight'] < 3000)]

# Find the most common origin in the filtered data
most_common_origin = filtered_cars['origin'].mode()[0]

print(f"The most common origin for cars with horsepower > 100 and weight < 3000 is: {most_common_origin}")
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
#your code here
mean_acceleration_japan = df[df['origin'] == 'japan']['acceleration'].mean()
mean_acceleration_japan_rounded = round(mean_acceleration_japan, 2)
print(f"Mean acceleration of cars from Japan: {mean_acceleration_japan_rounded}")
```

### 5. Which year had the highest average mpg?

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
# Calculate median mpg
median_mpg = df['mpg'].median()

# Filter cars with mpg above median
above_median_mpg = df[df['mpg'] > median_mpg].copy()

# Calculate horsepower-to-weight ratio
above_median_mpg['hp_per_weight'] = above_median_mpg['horsepower'] / above_median_mpg['weight']

# Find max hp_per_weight value
max_ratio = above_median_mpg['hp_per_weight'].max()

# Get all cars with this max ratio
best_cars = above_median_mpg[above_median_mpg['hp_per_weight'] == max_ratio]

print("Car(s) with best horsepower-to-weight ratio and above-median mpg:")
print(best_cars[['name', 'horsepower', 'weight', 'mpg', 'hp_per_weight']])
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
import pandas as pd

# Reload data to ensure 'name' is present
df = pd.read_csv("/content/Cars.csv")

# 1. Identify car names that appear in more than one model_year
name_year_counts = df.groupby('name')['model_year'].nunique()

# Filter names with 2 or more different years
multi_year_names = name_year_counts[name_year_counts >= 2].index

# 2. For each such name, compute std deviation and mean of mpg and count appearances
stats = (
    df[df['name'].isin(multi_year_names)]
    .groupby('name')
    .agg(
        mpg_std=('mpg', 'std'),
        mpg_mean=('mpg', 'mean'),
        appearances=('name', 'count')
    )
    .reset_index()
)

# 3. Filter those with std(mpg) < 1.0
consistent_cars = stats[stats['mpg_std'] < 1.0]

# 4. Find car(s) with the lowest variation (std) in mpg among filtered
min_std = consistent_cars['mpg_std'].min()
best_consistent_cars = consistent_cars[consistent_cars['mpg_std'] == min_std]

# Bonus: Sort by appearances (descending), then mpg_mean (descending)
best_consistent_cars = best_consistent_cars.sort_values(
    by=['appearances', 'mpg_mean'], ascending=[False, False]
).reset_index(drop=True)

print("Consistent car models with low mpg variation (<1.0 std):")
print(best_consistent_cars[['name', 'appearances', 'mpg_mean', 'mpg_std']])
```
